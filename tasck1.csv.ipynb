```python
df['actual_start_date'] = pd.to_datetime(df['actual_start_date'], errors='coerce')
df['activation_date'] = pd.to_datetime(df['activation_date'], errors='coerce')
time_difference = df['activation_date'] - df['actual_start_date']
average_days = time_difference.dt.days.abs().mean()
print(average_days)
```

```python
df['activation_date'] = pd.to_datetime(df['activation_date'], errors='coerce')
april_activations = df[df['activation_date'].dt.month == 4]
count_april = len(april_activations)
print(f"Total products activated in April: {count_april}")
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data.csv') 

df['is_active'] = np.where(df['current_status'] == 'Active', 1, 0)

active_counts = df.groupby('distribution_channel_value')['is_active'].sum().reset_index(name='Active_Count')

highest_active_channel = active_counts.sort_values(by='Active_Count', ascending=False).iloc[0]

print(f" {highest_active_channel['distribution_channel_value']}")
print(f" {highest_active_channel['Active_Count']}")
```
